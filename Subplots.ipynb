```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context='notebook', style='darkgrid', 
              palette='deep', font='sans-serif', 
              )
plt.rcParams['figure.dpi'] = 300
```

```python
df = pd.read_csv("breast_cancer.csv")
```

## 1*3 plot

```python
fig, axs = plt.subplots(1, 3, figsize=(16, 4))

sns.histplot(x = df['area_mean'], data=df, ax=axs[0])

sns.countplot(x = df['diagnosis'], data=df, ax=axs[1])

sns.scatterplot(x = df['concavity_mean'], y=df['area_mean'], data=df, ax=axs[2])

plt.show()
```

## 2*2 plot

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

sns.histplot(x = df['area_mean'], data=df, ax=axs[0][0])

sns.countplot(x = df['diagnosis'], data=df, ax=axs[0][1])

sns.scatterplot(x = df['concavity_mean'], y=df['area_mean'], data=df, ax=axs[1][0])

sns.scatterplot(x = df['area_mean'], y=df['texture_worst'], data=df, ax=axs[1][1])

plt.show()
```
